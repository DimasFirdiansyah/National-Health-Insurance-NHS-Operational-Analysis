### Exploring the Data

Through this exploration steps, by familiarising the available data could help to determine:
- Number of location in the data set
- Highest number of location in records
- service setting, context types, national categories, and appointment statuses

#### Number of Location in the data set

```python
# using value_count to count specific value in location column
# create variable to count location to eliminate duplication between other column values
count_locations = nc['sub_icb_location_name'].value_counts()
num_locations = count_locations.count()

# View number of location
print("Number of locations:", num_locations)
```

#### Highest number of location in records

```python
# find highest location in records
highest_loc_records= nc['sub_icb_location_name'].value_counts().nlargest(5)

# Create DataFrame for better presentation
highest_loc_df = pd.DataFrame({'Location': highest_loc_records.index, 'Count':highest_loc_records.values})

# make the index start from 1 for readability
highest_loc_df.index = range(1, len(highest_loc_df)+1)

# View the data
print('The 5 highest location in records :')
highest_loc_df
```

#### How many Service settings, context types, national categories, and appointment statuses are there?
##### Service setting

```python
# Find how many service settings in the data-sets
service_set_count = nc['service_setting'].value_counts()

# View the data
print('\033[1m'+'Number of service setting:'+'\033[0m')
print(service_set_count)
```

##### Context types

```python
# find how many context types
context_type_count = nc['context_type'].value_counts()

# View the data
print('\033[1m'+'Number of context types :'+'\033[0m')
print(context_type_count)
```

##### National Categories

```python
# find how many national categories in data set
nation_category_count = nc['national_category'].value_counts()

# view the data
print('\033[1m'+'Number of national category :'+'\033[0m')
print(nation_category_count)
```

##### Appointment Status

```python
# Find how many appointment status
appointment_stat_cnt = ar['appointment_status'].value_counts()

print(appointment_stat_cnt)
```
